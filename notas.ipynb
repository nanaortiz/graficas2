```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Datos 
data = pd.read_csv('datos.csv', sep='\t')
```

```python
data
```

```python
df_M = data[data['materia'] == 'Matemáticas']
df_M
```

```python
df_H = data[data['materia'] == 'Historia']
df_H
```

```python
df_C = data[data['materia'] == 'Ciencias']
df_C
```

```python
df_L = data[data['materia'] == 'Lenguaje']
df_L
```

```python
plt.boxplot([df_M['nota'], df_H['nota'], df_C['nota'], df_L['nota']], labels=['Matemáticas','Historia','Ciencias','Lenguaje'])
plt.title('Distrubución de notas')
plt.grid()
plt.ylabel("Nota")
plt.show()
```

2. Pie Chart

```python
Aprobado = data.groupby('aprobado').size()
etiquetas = 'No Aprobados', 'Aprobados'
colors = ['orangered', 'dodgerblue']
```

```python
Aprobado
```

```python
plt.pie(Aprobado, labels=etiquetas, autopct='%1.1f%%', startangle = 18, colors=colors)
plt.title('Distribución de Aprobados')
plt.show()
```
